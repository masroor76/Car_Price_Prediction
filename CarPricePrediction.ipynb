```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.preprocessing import LabelEncoder
import pickle
```

```python
dataset = pd.read_csv(r'car_price_dataset.csv')
dataset.head(10)
```

```python
df_coded = dataset
```

```python
le = LabelEncoder()
df_coded.Brand = le.fit_transform(df_coded.Brand) 
df_coded.Model = le.fit_transform(df_coded.Model) 
df_coded.Fuel_Type = le.fit_transform(df_coded.Fuel_Type) 
df_coded.Transmission = le.fit_transform(df_coded.Transmission) 
df_coded
```

```python
X = df_coded.iloc[:,:-1]
y = df_coded['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2)
```

```python
regression = linear_model.LinearRegression()
```

```python
regModel = regression.fit(X_train,y_train)
```

```python
regModel
```

```python
regModel.score(X_test,y_test)
```

```python
pred = regModel.predict(X_test)
print(pred)
```

```python
eff = regModel.score(X_test, y_test)
```

```python
efficiency = eff*100
efficiency
```

```python
with open('CarModel.pkl','wb') as f:
    pickle.dump(regModel,f)
```
